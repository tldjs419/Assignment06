01) 5 -> 3 -> 1 -> 8 -> 7 -> 6 -> 9

04) 정리(Theorem): 이진 검색 트리 T의 중위 순회는 노드들을 정렬된 순서로 방문한다.

<증명>
베이스 케이스: h = 1 (h: height)  
트리는 하나의 노드(루트)만 가진다.  
이 경우 하나의 원소만 존재하므로, 중위 순회 결과는 항상 정렬된 상태이다.

귀납적 가정:  
높이가 k < h 인 모든 이진 검색 트리에 대해, 중위 순회는 정렬된 순서로 방문한다고 가정한다.

귀납적 전개:  
높이가 h인 이진 검색 트리 T를 생각하자.  
이 트리는 루트 r, 왼쪽 서브트리 T_L, 오른쪽 서브트리 T_R로 구성된다.

이진 검색 트리의 성질에 의해  
T_L의 모든 키는 r보다 작고,  
T_R의 모든 키는 r보다 크다.

귀납적 가정에 의해  
T_L과 T_R의 중위 순회 결과는 각각 정렬되어 있다.

중위 순회는 T_L → r → T_R 순서로 방문하므로,  
전체 방문 순서는 정렬된 순서가 된다.

따라서 높이가 h인 경우에도 성립한다.

결론적으로 모든 이진 검색 트리에 대해 중위 순회는 항상 정렬된 순서를 따른다.

## 09

최대 높이는 트리가 한쪽으로만 치우친 경우이며, 루트를 높이 1로 볼 때 n개의 노드가 일렬로 연결된 형태가 된다. 따라서 최대 높이는 n이다. 최소 높이는 트리가 균형을 이룰 때이며, 루트를 높이 1로 볼 경우 ⌈log₂(n+1)⌉로 나타낼 수 있다.

최대 높이: n / 최소 높이: log₂(n+1)
